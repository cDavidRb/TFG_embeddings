```python
import spacy
import subprocess

try:
    # Cargar el modelo "en_core_web_sm"
    nlp = spacy.load("en_core_web_sm")
except OSError:
    # Si el modelo no está disponible, descargarlo usando subprocess
    subprocess.run("python -m spacy download en_core_web_sm", shell=True)
    nlp = spacy.load("en_core_web_sm")
```

```python
import re
#import nltk
#from nltk.corpus import stopwords
#from gensim.models import KeyedVectors

#embedding = KeyedVectors.load_word2vec_format("GoogleNews-vectors-negative300.bin", binary=True)


def limpiar(doc):
    
    doc = re.sub(r'\S+@\S+', '', doc) #Elimina las direcciones de correo
    doc = re.sub(r"[^a-zA-Z\s'-]", '', doc) #Elimina todos los caracteres no alfabéticos excepto los apóstrofes y guiones
    doc = re.sub(r'\s+', ' ', doc) #Convierte todos los espacios (tanto múltiples, tabulados y saltos de línea en " ")
    doc = doc.strip()
    
    doc = nlp(doc)
    
    doc = [token.lemma_.lower() for token in doc]
    
    doc = [palabra for palabra in doc if palabra not in stopwords]
    #final = [re.sub(r'[^a-zA-Z\s]', '', word) for word in sin_stopwords] #Elimina todos los caracteres no alfabéticos
    
    #en_ingles = [palabra for palabra in sin_stopwords if palabra in embedding]
    
    #doc_final = ' '.join(final)
    
    #final = [re.sub(r"'\s*s\s*", '', word) for word in sin_stopwords]
    #final = [re.sub(r"\b\w*'\w*\b", '', word) for word in final]
    
    return ' '.join([word for word in doc if len(word)>2])
```
